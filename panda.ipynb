```python
#A  method to clean and restructure the Edition column

def splitter(data):
    EditionType=[]
    for item in data:
        obj = item.split(",– ")[0].strip().upper()
        EditionType.append(obj)
    EditionMonth=[]
    EditionYear=[]
    months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
    for item in data:
        obj = item.split(",– ")[1].strip()
        obj = obj.split()[-2:]
        EditionMonth.append(obj[0].upper() if obj[0].upper() in months else 'NONE')
        if len(obj) > 1:
            if obj[1][0] >= '0' and obj[1][0] <= '9':
                EditionYear.append(int(obj[1]))
            else:
                EditionYear.append(2008)
        else:
            if obj[0][0] >= '0' and obj[0][0] <= '9':
                EditionYear.append(int(obj[0]))
            else:
                EditionYear.append(2008)
    return EditionType,EditionMonth,EditionYear
```

```python
# split authors
all_authors = []
def author_splitter(data):
    
    Author1 = []
    Author2 = []
    Author3 = []
    Author4 = []
    Author5 = []
    Author6 = []
    Author7 = []
    
    for item in data:
        item = item.split(",")
        
        try:
            Author1.append(item[0].upper())
            all_authors.append(item[0].upper())
        except:
            Author1.append('NONE')
            
        try:
            Author2.append(item[1].upper())
            all_authors.append(item[1].upper())
        except:
            Author2.append('NONE')
            
        try:
            Author3.append(item[2].upper())
            all_authors.append(item[2].upper())
        except:
            Author3.append('NONE')
            
        try:
            Author4.append(item[3].upper())
            all_authors.append(item[3].upper())
        except:
            Author4.append('NONE')
            
        try:
            Author5.append(item[4].upper())
            all_authors.append(item[4].upper())
        except:
            Author5.append('NONE')
            
        try:
            Author6.append(item[5].upper())
            all_authors.append(item[5].upper())
        except:
            Author6.append('NONE')
            
        try:
            Author7.append(item[6].upper())
            all_authors.append(item[6].upper())
        except:
            Author7.append('NONE')
    return Author1,Author2,Author3,Author4,Author5,Author6,Author7
all_authors.append('NONE')
```

```python
# to split the genre
# genre = db_train['Genre']
# maxi=1
# for item in genre:
#     item = item.split('&')
#     print(item)
#     if len(item)  > maxi:
#         maxi=len(item)
# print(maxi)
all_genre = []
def genre_splitter(data):
    G1 = []
    G2 = []
    for item in data:
        item = item.split(' & ')
        if len(item) == 1:
            G1.append(item[0].upper())
            all_genre.append(item[0].upper())
            G2.append('NONE')
        else:
            G1.append(item[0].upper())
            all_genre.append(item[0].upper())
            G2.append(item[1].split(' (')[0].strip().upper())
            all_genre.append(item[1].split(' (')[0].strip().upper())
    return G1,G2
all_genre.append('NONE')
```

```python
# to split the category 
all_category = []
def category_splitter(data):
    C1 = []
    C2 = []
    for item in data:
        item = item.split(",")
        if len(item) == 1:
            C1.append(item[0].strip().upper())
            all_category.append(item[0].strip().upper())
            C2.append('NONE')
        else:
            all_category.append(item[0].strip().upper())
            all_category.append(item[1].strip().upper())
            C1.append(item[0].strip().upper())
            C2.append(item[1].strip().upper())
    return C1,C2
all_category.append('NONE')
```

```python
def uppertitle(data):
    titles = list(data)
    titles = [title.strip().upper() for title in titles]
    return titles    
```

```python
def fetchreviews(data):
    results = []
    for item in data:
        item = item.split(" ")[0].strip()
        numb=0
        for i in range(len(item)):
            if item[i] >= '0' and item[i] <= '9':
                num = int(item[i])
                numb = numb * 10
                numb = numb + num
        results.append(int(numb))
    return results
fetchreviews(db_train['Ratings'].head(3))
def fetchratings(data):
    results = []
    for item in data:
        item = item.split(" ")
        rating = float(item[0])
        ratedon = float(item[3])
        expo = (5.0 * rating) / ratedon
        results.append(expo)
    return results
fetchratings(db_train['Reviews'].head(3))
```

```python
def clean(data):
    Titles = uppertitle(data['Title'])
    A1,A2,A3,A4,A5,A6,A7 = author_splitter(data['Author'])
    EditionType , EditionMonth , EditionYear = splitter(data['Edition'])
    Reviews = fetchreviews(data['Ratings'])
    Ratings = fetchratings(data['Reviews'])
    Genre1 , Genre2 = genre_splitter(data['Genre'])
    BookCategory1,BookCategory2 = category_splitter(data['BookCategory'])
    cleaned_data = pd.DataFrame({'Title':Titles,
                                 'Author1':A1,
                                 'Author2':A2,
                                 'Author3':A3,
                                 'Author4':A4,
                                 'Author5':A5,
                                 'Author6':A6,
                                 'Author7':A7,
                                 'Edition_Type':EditionType,
                                 'Edition_Month':EditionMonth,
                                 'Edition_Year':EditionYear,
                                 'Reviews':Reviews,
                                 'Ratings':Ratings,
                                 'Genre1':Genre1,
                                 'Genre2':Genre2,
                                 'Book_Category1':BookCategory1,
                                 'Book_Category2':BookCategory2,
                                })
    return cleaned_data

# db_train = clean(db_train)
```

```python

# A method for Finding Unique items for all columns
def unique(list1, list2):
  a = list1
  b = list2
  a.extend(b)
  return list(set(a))
```
